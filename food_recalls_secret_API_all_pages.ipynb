### what does extend do? explanation soma:


```python
#now we have to collect that into a single list
#we will use extend it combines multiple elements into one list
#my headliness lives in data['entries']

count = 1
headlines = []

for page_num in range(33):
    page = page_num + count
    print('-------')
    url = f"https://www.recallswiss.admin.ch/customer-access-backend/resources/recalls?page={page}"
    print("Requesting", url)
    response = requests.get(url)#just to show me what it does and if there is a problem somewhere
    data = response.json()
    headlines.extend(data['entries'])
```
